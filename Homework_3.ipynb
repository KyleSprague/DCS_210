```r

ggplot(data = edibnb, mapping = aes(x = price)) + 
  geom_histogram(binwidth = 40) + 
    facet_wrap(~neighbourhood,
    labeller = label_wrap_gen(width = 12), 
    dir = "h")
```

##ANSWER TO QUESTION 3 

I chose to have the facets ordered in the horizontal direction with price along the x-axis and the count of rooms along the y-axis because I felt it was easier to compare the extent of rooms in different areas which fall in the same price range or different price range.

```r
##code for question 4 part 1 

edibnb %>%  
  drop_na() %>%
  group_by(neighbourhood) %>%
  summarise(median_price = 
  median(price)) %>%
  arrange(desc(median_price)) %>%
  slice_max(median_price, n=5) 


  


```

```r
##code for question 4 part 2 

edibnb %>%  
  filter(neighbourhood == "New Town" |
  neighbourhood == "Old Town" | 
  neighbourhood == "West End" | 
  neighbourhood == "Stockbridge" | 
  neighbourhood == "Southside") %>%

  ggplot(aes(x = price, y = neighbourhood, fill = neighbourhood)) + 
    geom_density_ridges() 
    
  
#plot code inspired by: https://r-graph-gallery.com/294-basic-ridgeline-plot.html
```

```r
##Code for question 4, part 3

edibnb %>%  
  drop_na() %>% 
  
  filter(neighbourhood == "New Town" |
  neighbourhood == "Old Town" | 
  neighbourhood == "West End" | 
  neighbourhood == "Stockbridge" | 
  neighbourhood == "Southside") %>% 

  group_by(neighbourhood) %>% 
  summarize(
    min_price = min(price), 
    mean_price = mean(price), 
    median_price = median(price), 
    stdev_price = sd(price), 
    price_IQR = IQR(price)
  )


```
