##ANSWER TO QUESTION 5

In scatter plots with many individual points, as we have in our data set, information can get lost when points overlap. Jitter allows us to slightly shift the data points to make the extent to which a specific observation occurs at a specific point more obvious. For instance, in the data above, we see a smaller discrepancy between men and women's thumb sizes when we can see all the data points as a consequence of the imposed jitter. While with just the point plot it looked as though there was a correlation between sex and thumb size, with jitter the correlation seems less obvious. Further, we see how skewed the data is; there are many more female respondents to the survey than males, making it harder to conclude whether or not thumb length shows a correlation with sex.

##ANSWER TO QUESTION 10: 

A good explanatory variable is something that explains changes in the response variable, which implies that there is a correlation between the two. Correlations can not be deduced from boxplots (only distriutions can be) (Source: https://www.kaggle.com/questions-and-answers/92709). However, we can get correlation from a scatter plot. For sex and thumb length, we see that the proportion of the response variable (thumb length) that can be explained by sex is R^2 = 0.11, which is very small. This implies that sex is a poor explanatory variable for thumb length. More poor, however, is class year, the chosen explanatory variable. For this R^2 is 0.0045; in a practical sense, this implies that there is no relationship between class year and thumb length. This is to be expected given that, in university, most people have reached their full size. Therefore, year should not influence thumb length size much when compared to sex. It is worth noting that there is only one fourth year student in the data set, which slightly skews the data. However, since there was almost no correlation between the other three years, its effect on the least squares regression line is negligible.
