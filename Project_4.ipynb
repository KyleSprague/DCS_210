##ANSWER TO QUESTION 5

In scatter plots with many individual points, as we have in our data set, information can get lost when points overlap. Jitter allows us to slightly shift the data points to make the extent to which a specific observation occurs at a specific point more obvious. For instance, in the data above, we see a smaller discrepancy between men and women's thumb sizes when we can see all the data points as a consequence of the imposed jitter. While with just the point plot it looked as though there was a correlation between sex and thumb size, with jitter the correlation seems less obvious. Further, we see how skewed the data is; there are many more female respondents to the survey than males, making it harder to conclude whether or not thumb length shows a correlation with sex. 

##CODE FOR QUESTION 6 BELOW

```r
gf_jitter(Thumb ~ Sex, data = Fingers, color ~ Sex == "male", alpha = 0.5) %>%  
gf_refine(scale_color_manual(name = 'Male and Female', values = setNames(c('#d95foe','#2c7fb8'),c(T, F)))) %>% 
    gf_theme(theme_minimal()) %>%
  gf_labs(title = "Thumb Length (mm) and Individuals of a Particular Sex",
    y= "Thumb Length (mm)", x = "Participant Sex") %>% 
  gf_theme(plot.title = element_text(size= 18, hjust = 0.4), 
    axis.text = element_text(size = 12), 
  axis.title = element_text(size = 14))  

#Fingers %>%
#  ggplot(aes(x = Thumb, y = Sex)) +
#  geom_point() + 
#  geom_jitter(alpha = 0.5) +
# labs(
#  title = "Thumb Length (mm) Versus Sex for the Fingers Dataset",
#  subtitle = "Includes Added Jitter", 
#  ) +
#  labs(x= "Thumb Length (mm)", y = "Participant Sex") + 
#  theme_minimal() 
```

##ANSWER TO QUESTION 6

Adding an alpha value to a jitteer plot adds opacity to the points. It reveals to us the effect of the jitter. If points are very extensive in their overlap, we expect there to still be overlap despite a jitter. We see a little bit of this in the female data at a thumbl length of 55mm whereas we see almost no overlap in the male data. 

```r
gf_boxplot(Thumb ~ Sex, data = Fingers) %>% 
gf_refine(scale_color_manual(values = c(male = "#d95foe", female = "#2c7fb8"))) %>% 
    gf_theme(theme_minimal()) %>%
  gf_labs(title = "Thumb Length (mm) and Individuals of a Particular Sex",
    y= "Thumb Length (mm)", x = "Participant Sex") %>% 
  gf_theme(plot.title = element_text(size= 18, hjust = 0.4), 
    axis.text = element_text(size = 12), 
  axis.title = element_text(size = 14), 
  axis.title.x = element_text(vjust = -1),
  axis.title.y = element_text(vjust = 1)) 
```
